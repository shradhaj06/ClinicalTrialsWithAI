```python
import pandas as pd

# Load the dataset
data = pd.read_csv('clin_trials.csv')

# Display the first few rows of the dataset
print(data.head())

# Identify numerical and categorical columns
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = data.select_dtypes(include=['object']).columns

# Fill missing values
data[numerical_cols] = data[numerical_cols].fillna(-1)  # Replace numerical NaNs with -1
data[categorical_cols] = data[categorical_cols].fillna('Unknown')  # Replace categorical NaNs with 'Unknown'

# Display the updated dataset info
print(data.info())
```

```python
# Convert date columns to datetime
data['Start Date'] = pd.to_datetime(data['Start Date'], errors='coerce')
```

```python
import pandas as pd

# Load the dataset (ensure this is a DataFrame)
data = pd.read_csv('clin_trials.csv')  # Replace with your actual file path

# Check the type of 'data'
print(type(data))  # Should be <class 'pandas.core.frame.DataFrame'>

# Display the first few rows of the DataFrame
print(data.head())

# Check the columns
print(data.columns)

# Define numerical and categorical columns
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_cols = data.select_dtypes(include=['object']).columns.tolist()

# Display the updated lists of columns
print("Numerical Columns:", numerical_cols)
print("Categorical Columns:", categorical_cols)

# Descriptive statistics for numerical columns
if numerical_cols:  # Check if there are any numerical columns
    print(data[numerical_cols].describe())
else:
    print("No numerical columns found.")

# Descriptive statistics for categorical columns
if categorical_cols:  # Check if there are any categorical columns
    print(data[categorical_cols].describe())
else:
    print("No categorical columns found.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Plotting Overall Status
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Overall Status', order=data['Overall Status'].value_counts().index)
plt.title('Distribution of Overall Status')
plt.xticks(rotation=45)
plt.show()

# Plotting Organization Class
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Organization Class', order=data['Organization Class'].value_counts().index)
plt.title('Distribution of Organization Class')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting the distribution of Standard Age
plt.figure(figsize=(10, 6))
sns.histplot(data['Standard Age'], bins=10, kde=True)
plt.title('Distribution of Standard Age')
plt.xlabel('Standard Age (Encoded)')
plt.ylabel('Frequency')
plt.show()
```

```python

# Box plots for numerical columns
for col in numerical_cols:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=data[col])
    plt.title(f'Box Plot of {col}')
    plt.show()
```

```python
# Correlation matrix
correlation_matrix = data[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pair plot for numerical columns
sns.pairplot(data[numerical_cols])
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

```python
# Count plots for categorical variables
categorical_cols = data.select_dtypes(include=['object']).columns

for col in categorical_cols:
    plt.figure(figsize=(10, 5))
    sns.countplot(y=data[col], order=data[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Group analysis example
grouped_data = data.groupby('Overall Status')[numerical_cols].mean()
print(grouped_data)
```
